```python
import numpy as np
import pandas as pd
```

```python
match = pd.read_csv("match.csv")
match.head()
```

```python
match = match.drop(['home_player_1', 'home_player_2', 'home_player_3', 'home_player_4', 'home_player_5', 'home_player_6', 'home_player_7', 'home_player_8', 'home_player_9', 'home_player_10', 'home_player_11'], axis=1)
match = match.drop(['away_player_1', 'away_player_2', 'away_player_3', 'away_player_4', 'away_player_5', 'away_player_6', 'away_player_7', 'away_player_8', 'away_player_9', 'away_player_10', 'away_player_11'], axis=1)
match.shape
```

```python
def change_date_type(row):
    date = row.date.split('/')
    if (date[0] != "10" and date[0] != "11" and date[0] != '12'):
        date[0] = '0' + date[0]
    date = [date[2][:2], date[0], date[1]]
    return '/'.join(date)
match['date'] = match.apply(change_date_type, axis=1)
match.head(15)
```

```python
leagues = match.league_id.unique()
seasons = match.season.unique()
print(leagues)
print(seasons)
```

```python
def select_competition(matches, league, season):
    return matches.loc[(matches['league_id'] == league) & (matches['season'] == season)]
```

```python
a_competition = select_competition(match, 1, '2008/2009')
```

```python
def calculate_net_goal(a_competition):
    net_score_dict = {}
    teams = set(a_competition.home_team_id.unique().tolist() + a_competition.away_team_id.unique().tolist())
    for team in teams:
        net_score_dict[team] = 0
    
    home_net_score = []
    away_net_score = []
    a_competition = a_competition.sort_values(by=['date'], axis=0)
    for i in range(a_competition.shape[0]):
        home_net_score.append(net_score_dict[a_competition.home_team_id.iloc[i]])
        away_net_score.append(net_score_dict[a_competition.away_team_id.iloc[i]])
        net_score_dict[a_competition.home_team_id.iloc[i]] += (a_competition.home_team_goal.iloc[i] - a_competition.away_team_goal.iloc[i])
        net_score_dict[a_competition.away_team_id.iloc[i]] += (a_competition.away_team_goal.iloc[i] - a_competition.home_team_goal.iloc[i])
    a_competition.insert(a_competition.shape[1], 'home_net_score', home_net_score)
    a_competition.insert(a_competition.shape[1], 'away_net_score', away_net_score)
    return a_competition
```

```python
datas = []
for league in leagues:
    for season in seasons:
        competition = select_competition(match, league, season)
        competition = calculate_net_goal(competition)
        datas.append(competition)
match = pd.concat(datas, sort=False)
match.head(20)
```

```python
match = match.fillna(0)
match.isna().sum()
```

```python
def find_average_bet_H(match):
    numb_available = 0
    total = 0
    l = ['B365H', 'BWH', 'IWH', 'LBH', 'PSH', 'WHH', 'SJH', 'GBH', 'BSH', 'VCH']
    for col in l:
        if match[col] != 0:
            numb_available += 1
            total += match[col]
    if numb_available == 0:
        return 0
    return total/numb_available
def find_average_bet_D(match):
    numb_available = 0
    total = 0
    l = ['B365D', 'BWD', 'IWD', 'LBD', 'PSD', 'WHD', 'SJD', 'GBD', 'BSD','VCD']
    for col in l:
        if match[col] != 0:
            numb_available += 1
            total += match[col]
    if numb_available == 0:
        return 0
    return total/numb_available
def find_average_bet_A(match):
    numb_available = 0
    total = 0
    l = ['B365A', 'BWA', 'IWA', 'LBA', 'PSA', 'WHA', 'SJA', 'GBA', 'BSA', 'VCA']
    for col in l:
        if match[col] != 0:
            numb_available += 1
            total += match[col]
    if numb_available == 0:
        return 0
    return total/numb_available
```

```python
match['avg_h_bet'] = match.apply(find_average_bet_H, axis=1)
match['avg_d_bet'] = match.apply(find_average_bet_D, axis=1)
match['avg_a_bet'] = match.apply(find_average_bet_A, axis=1)
```
